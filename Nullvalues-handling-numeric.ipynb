```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
df=pd.read_csv("data.csv")
```

```python
df
```

## 1. Removing rows

```python
df.dropna()
```

## 2. Imputers

```python
imputers=SimpleImputer(missing_values=np.nan,strategy="mean")
imputers.fit(df.iloc[:,1:3].values)
df.iloc[:,1:3]=imputers.transform(df.iloc[:,1:3].values)
```
