```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("googleplaystore.csv")
impute=SimpleImputer(missing_values=np.nan,strategy="mean")
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3]=impute.transform(df.iloc[:,2:3].values)
df=df.dropna()
```

```python
df.head()
```

## Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order wih respect to ratings

```python
df_new=df[(df["Category"]=="ART_AND_DESIGN")&(df["Rating"]>4.5)].sort_values(by="Rating",ascending=True)
df_new.head(10)
```


## Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in descending order wih respect to ratings

```python
df["Reviews"] = df["Reviews"].astype(int)
print(df["Reviews"].dtypes)
```

```python
df_new=df[(df["Category"]=="ART_AND_DESIGN")&(df["Rating"]>4.5)].sort_values(by="Reviews",ascending=False)
df_new.head(10)
```

## Which category is having the maximum average rating?

```python
df_new=df.groupby("Category")["Rating"].mean()
df_new.sort_values(ascending=False)
```

## How many apps are there in each category

```python
df_new=df.groupby("Category")["Type"].count()
df_new.sort_values(ascending=False)
```
