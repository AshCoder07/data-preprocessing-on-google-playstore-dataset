```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("data.csv")
```

```python
df
```

## 1.Imputers

```python
imputer=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
imputer.fit(df.iloc[:,0:4].values)
df.iloc[:,0:4]=imputer.transform(df.iloc[:,0:4].values)
```
